# Flow Control

### IF Condition

```python
# If condition is one of the flow control ways to control how code is executed based on some conditions
age = 25

if age > 18:                             # Before : we put a condition which result can only be True or False
  print('You are eligible to vote')      # After : we put the action we need to execute if only the condition is achieved (True)
```

```python
age = 16

if age >= 18 :                          # False

    print('You are eligibile to vote')

else :

    print('You are not eligibile to vote because you are under the legal age')
```

```python
# If condition using if, elif and else
grade = 60

if grade >= 85:              # If the grade is greater than 85, the condition will be true and 'Excellent' will be printed
  print('Excellent')

elif grade >= 75:            # If the grade is lower than 85, the first condition will be false, so it will go to the next elif and make the check again
  print('Very Good')

elif grade >= 65:            # If the grade is lower than 85 and 75, the first and second conditions will be false, so it will go to the next elif and make the check again
  print('Good')

else:                        # After all if and elif conditions were checked but none of them is achieved (True), then else will be executed
  print('Fair')
```

```python
# Nested IF Condition is to have an if condition inside another IF Condition

num = -3

if num % 2 == 0:                              # - First check if the numer is Even, if the condition is achieved (True), then whatever after (:) will be executed

  if num > 0:                                 # - After the Even condition was achieved, it will check now if number is positive or negative by a simple if else statement
    print('Number is Even and Positive')

  else:
    print('Number is Even and Negative')

else:                                         # - If the Even condition was not achieved (False), it will go directly to this else statement and execute whatever after (:)

  if num > 0:                                 # - After the even condition was not achieved (False), it will now check if number is positive or negative by a simple if else statement
    print('Number is Odd and Positive')

  else:
    print('Number is Odd and Negative')
```

### For Loop

```python
# For loop is a loop that repeates an order a specific number of times, like repeating the order of converting letters to upper case in a word

for letter in 'ahmed':
  print(letter.upper())
```

```python
# How for loop works ?
# The above code is just the same as the following, where the for loop variable is assigned each time to an element from the iterable we loop over

letter = 'a'
print(letter.upper())

letter = 'h'
print(letter.upper())

letter = 'm'
print(letter.upper())

letter = 'e'
print(letter.upper())

letter = 'd'
print(letter.upper())
```

```python
# For Loop is considered as one of the flow control like if condition and we can control the for loop using break and continue

# break: is used to stop the loop immediately

for letter in 'ahmed':
  if letter == 'd':
    break
  print(letter.upper())
```

```python
# continue: is used to skip an iteration and loop completes normally

for letter in 'ahmed':
  if letter == 'h':
    continue
  print(letter.upper())
```

```python
# else : is used to execute a specific order if the for loop was not interrupted by a break

for letter in 'ahmed':

  if letter == 'm':
    break

  print(letter.upper())

else:
  print('Done')
```

```python
name = 'ahmed'

for letter in name:

    if letter == 'm':
        print('Letter (m) was found')
        break

    print(letter)

else:
    print('Loop was completely done, Letter (m) was NOT found')
```

```python
name = 'ali'

for letter in name:

    if letter == 'm':
        print('Letter (m) was found')
        break

    print(letter)

else:
    print('Loop was completely done, Letter (m) was NOT found')
```

```python
# Nested For Loop

colors = ['red', 'green']
bags = ['bag_1', 'bag_2']

for color in colors:              # Outer for loop

    for bag in bags:              # Inner for loop
        print(color, bag)
```

### While Loop

```python
# While loop is similar to for loop in repeating orders, except that while loop repeats order based on a specific condition (True or False)

# In this example as long as the condition is True, the loop will work, once the condition becomes False when x = 5, the loop stops

x = 1
while x < 5:

    print(f'{x} is less than 5')

    x = x + 1
```

```python
# Like for loop we can also stop the while loop based on a specific condition in the code

name = 'ahmed'
id = 0
while id < len(name):

    if name[id] == 'e':
        break

    print(name[id])
    id = id + 1
```

```python
# Also to skip an iteration in the while loop we can use continue

name = 'ahmed'
id = 0
while id < len(name):

    if name[id] == 'e':    # Here we skip the iteration where the character is `e`, then proceed normlly the loop
        id = id + 1
        continue

    print(name[id])
    id = id + 1
```

```python
# else : is used to execute a specific order if the while loop was not interrupted by a break

name = 'ahmed'
id = 0
while id < len(name):

    if name[id] == 'm':
      print('Letter (m) was found')
      break

    print(name[id])
    id = id + 1

else:
    print('Loop was completely done, Letter (m) was NOT found')
```

### List Comprehension

```python
# List comprehension is a brief way to make a for a loop, then put the result in a list

my_list = []

for letter in 'ahmed':
  my_list.append(letter.upper())

print(my_list)
```

```python
# We can get the same result above by using List comprehension in a shorter way

my_list = [letter.upper() for letter in 'ahmed']        # we first open list brackets, then write the for loop, then write the order we need to repeat before the for loop

print(my_list)
```

```python
# Like Set, Dictionary elements are not ordered (Meaning it has no index)

my_dict = {'Name' : 'Mohamed', 'Age' : 20, 'Married' : True}
print(my_dict[0])
```

### How to get a dictionary value ?

```python
# To get a dictionary value, we call its key

my_dict = {'Name' : 'Mohamed', 'Age' : 20, 'Married' : True}
print(my_dict['Name'])                                        # Here we got the value 'Mohamed' by calling its key 'Name'
```

# Function

### Definition
#### As built-in functions and data types methods are considered functions because they do a specific operation, we can create our own functions and use it whenever we need as well.

**Advantages** **of** **using** **Functions** **:**
- Code Reusability: Functions allow us to reuse our code multiple times. This helps us save time and resources while coding.
- Readability: Functions make our code more readable. By breaking our code into small chunks, we can easily understand and debug our code.
- Abstraction: Functions provide abstraction. This means that we can hide the complexity of our code and focus on the task at hand.
- Maintainability: With functions, we can easily maintain our code. We can modify a single function without affecting the rest of the code.

```python
# Function Syntax

def welcome(name):                                # name : the variable we put inside the brackets is called `parameter` and we use it when the function requires this parameter to work

    '''This Function prints two welcome messages
    First messgae mention the name
    second message welcome to epsilon'''          # The string we write in the beginning of the function is the documentation of this function

    print(f'Hello {name}')
    print('Welcome to Epsilon')                   # After that we write our code normally that we created the function for

welcome('Mahmoud')                                # Now we use our function the same as any other built in function, and the Mahmoud is called `Argument` (The actual value of the parameter when we use the function)
```

### Print Vs Return

```python
# When we use print in a function to show the output of that function, it works well but we can't use the function in any operation after that
# Instead we use return which allows us to use the function after that as a variable or to make any operation on it normally
```

```python
# Using print

def add(a, b):
    print(a + b)

add(4, 5) * 2      # This error raises because the type of that function is the same type as print, which is NoneType and we can't do any operation on a NoneType
```

```python
# Using return

def add(a, b):
    return a + b

print(add(4, 5))
print(add(4, 5) * 2)     # Now we can make any operation normally on the function or deal with it as a variable
```

### Keyword Arguments

```python
# In all previous functions, we had to follow the order of parameters when we put the corresponding arguments
# Keyword Arguments allow us to use arguments without the neccessity of following the same orders of parameters

def full_name(first_name : str, mid_name : str, last_name : str):

    'Returns the full name of person'
    print(f'{first_name} {mid_name} {last_name}')

full_name(mid_name= 'Mohamed', first_name= 'Ahmed', last_name= 'Ali')        # Now regardless of the order we use, what matter is the argument we assign to each parameter
```

### Default Arguments

```python
# When we create a function with a specific number of parameters, it requires when we call the function to provide the number of arguments
# But if we have default arguments when creating the function, it's not a must in that case when we call it to provide all the arguemnts

def full_name(first_name = 'Ahmed', mid_name = 'Mohamed', last_name = 'Zaki'):    # Now we assigned default arguments to the parameters
    'Returns the full name of person'
    print(first_name + ' ' + mid_name + ' ' + last_name)
```

```python
full_name()                           # If we did not provide any arguments when calling the function, it will use the default ones
```

```python
full_name(first_name= 'Mohab', mid_name= 'Mohamed', last_name= 'Allam')  # If we used other arguments, it will update the default arguments
```

```python
full_name(first_name= 'Mohab', mid_name= 'Mohamed')  # If we only used some of the arguments, it will update the arguments provided and the others will be the default ones
```

### Arbitrary Arguments

```python
# If we need to create a function, but not sure when we use it how many arguments will be we provided (like function for summing numbers)

def avg(*args):                   # When we add `*` before the parameter, we define it as arbitrary parameter
    print(sum(args) / len(args))
```

```python
# Now when we use the function we can sum 2, 3, 4, 5, or any number of numbers
avg(10, 20, 30)
```

```python
avg(10, 20, 30, 40, 50, 60)
```

### Variable Scope

```python
# Variables we use in python normally are called global variables, because once created we can use them again and again

name = 'Ahmed'
print(name.upper())
```

```python
# Local variables are variables that are created within a function, and they are called local because we can not use them ou that function

def welcome():
    name = 'Ali'
    return name

print(welcome())

print(name)      # Notice here name variable was not updated because the second variable was a local variable inside a function
```

```python
# There is a way if we need to make a local variable a global one using built in `global`

def welcome():
    global name
    name = 'Ali'
    return name

print(welcome())

print(name)      # Now name variable was updated because the second variable was a global variable
```

### Lambda Expression

```python
# Lambda expression definition :
# A lambda expression is a small anonymous function defined using the lambda keyword.
# It can take any number of arguments, but can only have one expression.
# Lambda expressions are typically used for short, simple functions where a full def function definition would be overkill.

power_2 = lambda x : x ** 2

power_2(3)
```

```python

power_p = lambda x, p : x ** p

power_p(x= 3, p= 2)
```

### Handling Errors

#### try-except
#### The type of erros we can handle in python are Runtime errors and handling meaning we can proceed with our code normally after the line of code that returns an error
- Note: finally is not widely used because it always run, whether there is an error or not

<img src='https://files.realpython.com/media/try_except_else_finally.a7fac6c36c55.png'>

```python
# In this example we can handle many erros as per our inputs

try:

    l = [10, 20, 30, '40']
    index = int(input('Enter an Index for the list'))
    x = l[index]

    d = {'n1' : 0, 'n2' : 5, 'n3' : 10}
    key = input('Enter a key for the dictionary')
    y = d[key]
    print(x / y)

except ZeroDivisionError:
    print('Can not divide by zero')

except ValueError:
    print('Value Error')

except IndexError:
    print('Index Error')

except KeyError:
    print('Key Error')

except TypeError:
    print('Type Error')

else:
    print('No Errors Found')

print('Done')
```

```python
# In case we do not know the error but need to handle any error might result from the code

try:
    l = [10, 20, 30, '40']
    index = int(input('Enter an Index for the list'))
    x = l[index]

    d = {'n1' : 0, 'n2' : 5, 'n3' : 10}
    key = input('Enter a key for the dictionary')
    y = d[key]
    print(x / y)
except Exception as e:
    print(type(e))

print('Done')
```

### Random

```python
import random

# To get random integer between two numbers, including these two numbers
print(random.randint(1, 3))

# To # To get random integer between two numbers, including only start number number
print(random.randrange(1, 10))

# To get a random float number between 0 and 1
print(random.random())

# To get a random float number between two numbers
print(random.uniform(1, 3))
```

```python
# To get a random element from iterable (like: list, tuple, set, string)
print(random.choice([1, 2, 3, 4, 5]))

# To get more than random element from the list
print(random.choices([1, 2, 3, 4, 5], k=3))

# To get a random element from iterable but without duplication
print(random.sample([1, 2, 3, 4, 5], k=3))
```
